# NETFLIX Analysis Project

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

## Loading Dataset

```python
df = pd.read_csv('netflix1.csv')
df.head()
```

## Data Exploration

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

## Data Visualization 

```python
plt.figure(figsize =(8,6))
sns.countplot(x = "type",data =df, palette = "Set1")
plt.title("NETFLIX CONTENT TYPE");
```

```python
df["type"].value_counts()
```

```python
release20 = df[(df['release_year'] >= 2000 )]
```

```python
plt.figure(figsize =(14,8))
sns.countplot(data=release20,
            x= 'release_year',
            hue='type',
            palette = "Set1")

plt.title("NETFLIX CONTENT PRODUCTION IN 2000's");
```

```python
plt.figure(figsize = (10,6))
sns.countplot(x="country", 
              order = df["country"].value_counts().index[0:10],
              hue = "type",
              data = df,
             palette = "Set1")
plt.title("Top 10 Countries Production Countries On Netflix");
```

```python
plt.figure(figsize =(8,6))
df["rating"].value_counts().plot(kind ="line",
                                 color = 'black',
                                 linestyle = 'solid',
                                 marker = 'o',
                                 markerfacecolor = 'red',
                                 markersize = 7)   
plt.title("Netflix Rating Production");
```

```python
df["director"].value_counts().head(10)
```

```python
plt.figure(figsize=(15,9))
sns.countplot(x = "director",
              order = df["director"].value_counts().index[:10], 
              data = df)
plt.title("Top 10 Director");
```

Since missing value on **director** column is too huge, we cannot use this parameter to analyze.
